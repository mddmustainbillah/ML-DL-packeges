# Matplotlib Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Use Numpy to generate random data
"""
import numpy as np
x = np.linspace(0,10,25)
y = x*x + 2
print(x)
print(y)
```

```python
"""It only takes 1 command to draw
"""
pylab.plot(x,y, 'r') # 'r' stands for red
```

```python
"""Drawing a subgraph
"""
pylab.subplot(1,2,1) # The contents of the brackets represent(rows, columns, indexes)
pylab.plot(x,y, 'r--') # The third parameter here determines color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x,'g+-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.9, 0.8]) # Control the left, right ,width, height of the canvas (from 0 to 1)

axis.plot(x,y, 'r')
```

```python
# Again we can draw subgraphs
fig,axes = plt.subplots(nrows = 1, ncols = 2) # Submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# We can also draw a picture, or graph, inside another graph
fig = plt.figure()

# Control the left, right, width, height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Big axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Small axes

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize = (16,9), dpi = 300) # New graphic object

fig.add_subplot()

plt.plot(x,y, 'g')
```

```python
ax.legend(["label1", "label2"])

fig, axes = plt.subplots()
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title('Title of graph')

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(['y = x**2', 'y = x**3'], loc = 2)
```

```python
# In Matplotlib, you ca set other properties such as line color, transparency, and more.

fig, axes = plt.subplots(dpi = 150)
axes.plot(x, x**2, color = 'red', alpha = .5)
axes.plot(x, x+2, color = 'blue', alpha = .5)
axes.plot(x, x+3, color = 'green', alpha = .5)
```

```python
fig, ax = plt.subplots(dpi=100)

# Line width
ax.plot(x, x+1, color='blue', linewidth = 0.25)
ax.plot(x, x+2, color='blue', linewidth = 1)
ax.plot(x, x+3, color='blue', linewidth = 1.5)
ax.plot(x, x+4, color='blue', linewidth = 2)
```

```python
fig, ax = plt.subplots(dpi = 100)

# Line width, Line Style
ax.plot(x, x+1, color = 'blue', lw = 2, linestyle = '-')
ax.plot(x, x+2, color = 'blue', lw = 2, linestyle = '-.')
ax.plot(x, x+3, color = 'blue', lw = 2, linestyle = ':')

line, = ax.plot(x, x+4, color = 'black', lw = 1.50)
line.set_dashes([5, 10, 15, 10])

line, = ax.plot(x, x+5, color = 'black', lw = 1.50)
line.set_dashes([5, 10, 15, 3])

line, = ax.plot(x, x+6, color = 'black', lw = 1.50)
line.set_dashes([5, 10, 4, 10])
```

```python
fig, ax = plt.subplots(dpi = 100)

ax.plot(x, x+1, color = 'blue', marker = 'o')
ax.plot(x, x+2, color = 'blue', marker = '+')
ax.plot(x, x+3, color = 'blue', marker = 's')
ax.plot(x, x+4, color = 'blue', marker = '1')

ax.plot(x, x+5, color = 'blue', marker = 'o', markersize = 5)
ax.plot(x, x+6, color = 'blue', marker = 's', markerfacecolor = 'red')
ax.plot(x, x+7, color = 'blue', marker = 'o', markersize = 15, markerfacecolor = 'green')
```

```python
"""set the canvas grid and axis range
"""
fig, Axes = plt.subplots(1,2,figsize = (10,5))

Axes[0].plot(x, x**2, x, x**3, lw = 2)
Axes[0].grid(True)


Axes[1].plot(x, x**2, x, x**3, lw = 2)
Axes[1].set_ylim([0,60]) # limit values of y
Axes[1].set_xlim([2,5]) # limit values of x
Axes[1].grid(True)
```

# Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize = (16,5))
axes[0].set_title("Scatter")
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw = 2)

axes[2].set_title("bar")
axes[2].bar(n,n**2)

axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2, x**3, color = "green", alpha = 0.6)
```

```python
"""Draw a radar chart
"""
fig = plt.figure(figsize = (6,6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar = True)
t = np.linspace(0, 2**np.pi, 100)
ax.plot(t, t*2, color = 'blue', lw = 3)
```

```python
"""Draw a histogram
"""
n = np.random.randn(100000)
fig, axes = plt.subplots(1,2,figsize = (12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative = True, bins = 8)
```

# Draw contour image

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 -(Y-1)**2)
Z = (Z1 - Z2) * 2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title('Contour map')
```

# Draw 3D surface image

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize = (14,6))

# Specify the 3D graphics to draw, with projection = '3D'
ax = fig.add_subplot(1, 1, 1, projection = '3d')
ax.plot_surface(X,Y,Z, rstride = 4, cstride = 4, linewidth = 0)
```

# Pie chart of the popularity of programming languages

```python
import matplotlib.pyplot as plt
# Data to plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#','C++'
popularity = [22.2, 17.6, 8.8,8, 7.7, 6.7]
colors = ['red','green', 'blue','yellow','black','gray']
```

```python
# Explore 1st slice
explode = (0.1,0,0,0,0,0)
plt.pie(popularity, explode = explode, labels = languages, colors = colors,
       autopct ='%1.1f%%', shadow = True, startangle = 90)
```
